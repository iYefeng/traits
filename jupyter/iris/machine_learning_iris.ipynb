```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('data/iris.data', header=None, sep=',', names=['A', 'B', "C", "D", "label"])
label = df['label']
dataset = df[df.columns[0:4]]
label.head()
dataset.head()
```

```python
dfmax = dataset.max()
print dfmax
dfmin = dataset.min()
print dfmin
dflen = dfmax[0:4] - dfmin[0:4]
print dflen
dataset.head()
    
(dfmax-dfmin).T
```

```python
scaledDataSet = (dataset -(dfmax+dfmin)/2) / (dflen / 2)
```
